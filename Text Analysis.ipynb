## Importing Libraries

```python
import os
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk import word_tokenize
import re
import string
from string import punctuation
```

## Text Analysis Using NLP

```python
def Text_analysis(txt_file, output):
    print(txt_file)
    ### extracting text from files
    text=''
    filename = "G:/internshala project/Scraped articles/"+ txt_file
    with open(filename,'r', encoding='utf-8') as file:
        for i in file.readlines():
            text+= i
    text = text.replace("”"," ").replace("“"," ")
    text = " ".join(text.split())
    
    ### extracting stopwords from stopword directory
    def extract_stopword(path):
        stopword = []
        filename = "G:/internshala project/StopWords/" + path
        with open(filename,'r') as file:
            for i in file.readlines():
                if i.find("|"):
                    i = i[:i.find("|")].strip()
                    stopword.append(i.lower())
                else:
                    i = i.strip()
                    stopword.append(i.lower())
        return stopword  

    path = r"G:\internshala project\StopWords"
    dir_list = os.listdir(path)  
    
    Stopwords=[]
    for i in dir_list:
        Stopwords.extend(extract_stopword(i))
    Stopwords = list(set(Stopwords))
    print("Number of Stopwords: ",len(Stopwords))
    
    ### Tokenizing text inot words & Sentences
    sentences = sent_tokenize(text.lower())
    print("Sentence count in article :",len(sentences))
    
    words = word_tokenize(text.lower())
    print("Word count in article :",len(words))


    ### Removing extracted Stopwords from words
    word_w_stop = []
    for i in words:
        if i not in list(set(Stopwords)):
            word_w_stop.append(i)
    print("Word count after removing Stopwords :",len(word_w_stop))

    ### Extracting Negative & positive words from Master directory
    negative_words=[]
    filename = "G:/internshala project/MasterDictionary/negative-words.txt"
    with open(filename,'r') as file:
        for i in file.readlines():
            #print(i.strip().lower())
            negative_words.append(i.strip().lower())
    print("No. of Negative sentiment words: ",len(negative_words))

    positive_words=[]
    filename = "G:/internshala project/MasterDictionary/positive-words.txt"
    with open(filename,'r') as file:
        for i in file.readlines():
            #print(i.strip().lower())
            positive_words.append(i.strip().lower())
    print("No. of positive sentiment words: ",len(positive_words))
    
    ### Extracting Positive scores
    positive_score = 0
    for i in word_w_stop:
        if i in positive_words:
            positive_score+=1
    print("Positive Score: ",positive_score)
    
    ### Extracting Negative scores
    negative_score = 0
    for i in word_w_stop:
        if i in negative_words:
            negative_score-=1
    negative_score = -1*negative_score
    print("Negative Score: ",negative_score)  
    
    ### Getting Polarity Score using negative & positive score
    polarity_score = np.round((positive_score - negative_score)/ ((positive_score + negative_score) + 0.000001),3)
    print("Polarity Score: ",polarity_score)
    
    ### Word count after removing stopwords & Punctuations
    word_without_stop = []
    for i in words:
        if i not in set(stopwords.words('english')):
            word_without_stop.append(i)
    print("Words without Stop word: ",len(word_without_stop))

    cleaned_words = []
    
    for i in word_without_stop:
        if i not in punctuation:
            cleaned_words.append(i)
    word_count = len(cleaned_words)
    print("Total Cleaned Words: ", word_count)
    
    ### Total words after cleaning
    #Total_words_after_cleaning = len(word_w_stop)
    subjectivity_score = np.round((positive_score + negative_score)/(word_count + 0.000001),3)
    print("words after cleaning: ",word_count)
    print("Subjectivity Score: ",subjectivity_score)

    ### Average Number of words per Sentence
    Average_number_of_words_per_sentence = np.round((len(words)/len(sentences)),3)
    print("Average Number of Words Per Sentence: ",Average_number_of_words_per_sentence)
    
    ### Average Sentence Length
    Average_sentence_length = np.round((len(words)/len(sentences)),3)
    print("Average Sentence Length: ", Average_sentence_length)

    ### Syllables count per words
    def count_syllables(word):
        if word.endswith(("es", "ed")):
            return 0

        vowel_pattern = re.compile(r"[a,e, i, o, u]", re.IGNORECASE)
        vowel_count = len(vowel_pattern.findall(word))
        return max(1, vowel_count)
    
    ### calling Count_syllable function & getting Syllables count per word & Complex word count
    total_syllables = 0
    total_words = len(words)
    complex_word_count=0
    for word in words:
        total_syllables += count_syllables(word)
        ### complex words has greater than 2 syllables
        
        if count_syllables(word)>2:
            complex_word_count+=1
        average_syllables_per_word = np.round((total_syllables / total_words),3)
        
    print("Total Syllables: ",total_syllables)
    print("Syllables count per word: ",average_syllables_per_word)
    print("Complex word Count: ",complex_word_count)
    
    ### Calculating Percentage of complex words
    Percentage_of_Complex_words = np.round((complex_word_count / total_words),3) 
    print("%_of_Complex_words: ", Percentage_of_Complex_words)
    
    ### Calculating Fog Index to get Readability Score 
    Fog_Index = np.round(0.4 * (Average_sentence_length + Percentage_of_Complex_words),3)
    print("Fog_Index: ",Fog_Index)

    ### Personal Pronouns count
    def count_personal_pronouns(text):
        personal_pronouns = ["i", "we", "my", "ours", "us"]
        pattern = r'\b(?:' + '|'.join(personal_pronouns) + r')\b'
        # Use regex to find and count occurrences
        matches = re.findall(pattern, text, flags=re.IGNORECASE)
        count = len(matches)
        return count
    
    ### calling count_personal_pronoun function
    personal_pronoun_count = count_personal_pronouns(text)

    print("Personal pronouns count: ",personal_pronoun_count)
    
    ### Average Word Length
    count=0
    for each_word in words:
        for char in each_word:
            count+=1
    total_char = count
    Average_word_length = np.round(total_char/len(words),3)
    print("Average_Word_length: ",(Average_word_length))
    print(50*"--")
    
    
    output.append({'URL_ID': txt_file[:-4], 'POSITIVE SCORE':positive_score, 'NEGATIVE SCORE':negative_score, 'POLARITY SCORE':polarity_score,
                   'SUBJECTIVITY SCORE':subjectivity_score, 'AVG SENTENCE LENGTH': Average_sentence_length,
                   'PERCENTAGE OF COMPLEX WORDS': Percentage_of_Complex_words, 'FOG INDEX':Fog_Index, 'AVG NUMBER OF WORDS PER SENTENCE': Average_number_of_words_per_sentence,
                   'COMPLEX WORD COUNT': complex_word_count, 'WORD COUNT':word_count, 'SYLLABLE PER WORD': average_syllables_per_word,
                   'PERSONAL PRONOUNS': personal_pronoun_count, 'AVG WORD LENGTH': Average_word_length})
    return output
    
```

```python
import os 
path = r"G:\internshala project\Scraped articles"
files = os.listdir(path)
output=[]
for filename in files:
    output = Text_analysis(filename,output)
    
```
